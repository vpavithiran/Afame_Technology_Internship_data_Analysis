### Task 1 Sales dataset

```python
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/kaggle/input/sales-data-analysis/Sales Data.csv'
df = pd.read_csv(file_path)
```

### Droping the first feature 'Unnamed: 0'. Since it has no Use.

```python
df.drop('Unnamed: 0', axis=1 ,inplace=True)
```

### Null value check

```python
# missing value check
df.isnull().sum()
```

### Type Conversion

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df.info()
```

### Dropping Duplicate values

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

### Encode Categorical Variables


```python
# Get a list of categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Example: One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

### Normalize/Standardize Numerical Features(coping data to new dataframe(data))


```python
from sklearn.preprocessing import StandardScaler
data = pd.DataFrame(df)
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
```

```python
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

### Distribution of Sales

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

### Sales Over Time


```python
# Plot sales over time
plt.figure(figsize=(14, 7))
df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().plot()
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

### Top 10 Products by Sales


```python
# Top 10 products by sales
top_products = df.groupby('Product')['Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

### Sales by Region

```python
# Sales by region
region_sales = df.groupby('City')['Sales'].sum()
plt.figure(figsize=(12, 6))
region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

### Sales by Product

```python
# Sales by category
category_sales = df.groupby('Product')['Sales'].sum()
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

### Sales Trends by Product


```python
plt.figure(figsize=(14, 7))
for category in df['Product'].unique():
    df_category = df[df['Product'] == category]
    df_category.groupby(df_category['Order Date'].dt.to_period('M'))['Sales'].sum().plot(label=category)
plt.title('Sales Trends by Product')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### Correlation Analysis

```python
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### Pairplot Analysis

```python
sns.pairplot(df)
plt.show()
```

### Sales by Month

```python
# Sales by month
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### Boxplot of Sales by City

```python
# Boxplot of sales by City
plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='Sales', data=df)
plt.title('Boxplot of Sales by Region')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```

### Top 10 Customers(ID) by Sales


```python
# Top 10 customers by sales
top_customers = df.groupby('Order ID')['Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_customers.plot(kind='bar')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.show()
```

### Average Sales per Transaction

```python
df['Sales'].mean()
```

### Median Sales per Transaction

```python
df['Sales'].median()
```

### Sales Performance by Quarter


```python
# Sales performance by quarter
df['Quarter'] = df['Order Date'].dt.to_period('Q')
quarterly_sales = df.groupby('Quarter')['Sales'].sum()
plt.figure(figsize=(12, 6))
quarterly_sales.plot(kind='bar')
plt.title('Sales Performance by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.show()
```

### Sales Growth Rate by Month

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_growth_rate = monthly_sales.pct_change().fillna(0)
plt.figure(figsize=(12, 6))
monthly_growth_rate.plot()
plt.title('Monthly Sales Growth Rate')
plt.xlabel('Month')
plt.ylabel('Growth Rate')
plt.show()
```

### Seasonal Analysis of Sales


```python
seasonal_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(12, 6))
seasonal_sales.plot(kind='line')
plt.title('Seasonal Analysis of Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
